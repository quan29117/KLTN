#I. Read data

```python
import wfdb
import pandas as pd
from pathlib import Path

REQUIRED_SIGNALS = ['ECG', 'EMG', 'foot GSR', 'hand GSR', 'HR', 'RESP', 'marker']

def validate_drive(available_signals):
    available_lower = [s.lower() for s in available_signals]
    missing = []

    for req in REQUIRED_SIGNALS:
        found = any(req.lower() in s for s in available_lower)
        if not found:
            missing.append(req)

    return len(missing) == 0, missing

def read_data(input_path="./DRIVEDB"):
    base_path = Path(input_path).resolve()

    output_dir = base_path / "drive_csv"
    output_dir.mkdir(parents=True, exist_ok=True)

    dat_files = sorted(list(base_path.glob("*.dat")))

    for dat_file in dat_files:
        try:
            record_path = str(dat_file.with_suffix(''))

            signals, fields = wfdb.rdsamp(record_path)
            sig_names = fields['sig_name']

            is_valid, missing_sigs = validate_drive(sig_names)

            if not is_valid:
                print(f" Skipped {dat_file.name}: Missing {missing_sigs}")
                continue

            units = fields['units']
            col_names = [f"{sig}-{unit}".replace(",", "_") for sig, unit in zip(sig_names, units)]

            df = pd.DataFrame(signals, columns=col_names)

            save_path = output_dir / f"{dat_file.stem}.csv"
            df.to_csv(save_path, index=False)

            print(f"{dat_file.name} -> {save_path.name}")

        except Exception as e:
            print(f"Error processing {dat_file.name}: {e}")

read_data(input_path="./DRIVEDB")
```

## Labeling

```python
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt

RAW_CSV_DIR = Path("./DRIVEDB/drive_csv")
MARKER_INFO = Path("./DRIVEDB/marker_info_raw.csv")
LABELED_DIR = Path("./labeled_output")
LABELED_DIR.mkdir(parents=True, exist_ok=True)

FS = 15.5
WINDOW_SEC = 60
STEP_SEC = 30
MIN_SEG_SEC = 1

ECG_COL = "ECG-mV"
FGSR_COL = "foot GSR-mV"
HGSR_COL = "hand GSR-mV"
RESP_COL = "RESP-mV"
HR_COL = "HR-bpm"

SEG_ORDER = ["Rest1","City1","Hwy1","Return","Hwy2","City2","Rest2"]
RAW_TO_SHORT = {"Rest1":"rest","City1":"city","Hwy1":"highway","Return":"return",
                "Hwy2":"highway","City2":"city","Rest2":"rest"}

BINARY_MAP = {"rest":"no","city":"yes","highway":"yes","return":"yes"}
THREE_MAP  = {"rest":"low","city":"high","highway":"medium","return":"medium"}
```

```python
def load_marker_info(path):
    df = pd.read_csv(path)
    df.columns = [c.strip() for c in df.columns]
    df["Driver"] = df["Driver"].astype(str).str.strip().str.lower()

    for c in SEG_ORDER:
        if c not in df.columns:
            df[c] = np.nan
        else:
            df[c] = pd.to_numeric(df[c], errors="coerce")

    if "Start" in df.columns:
        df["Start"] = pd.to_numeric(df["Start"], errors="coerce").fillna(0)
    else:
        df.insert(1, "Start", 0)

    cols = ["Driver","Start"] + SEG_ORDER
    return df[cols]

marker_df = load_marker_info(MARKER_INFO)
marker_df
```

```python
import pandas as pd
import numpy as np
from scipy.signal import find_peaks
import matplotlib.pyplot as plt

def detect_marker_peaks(csv_path, column_name='marker-mV', distance_min=10, prominence_min=0.1, show_plot=False):
    df = pd.read_csv(csv_path)

    signal = df[column_name].values

    peak_indices, properties = find_peaks(
        signal,
        distance=distance_min,
        prominence=prominence_min
    )

    # 3. Visualization (Tùy chọn)
    if show_plot:
        plt.figure(figsize=(12, 4))
        plt.plot(signal, label='Signal')
        plt.plot(peak_indices, signal[peak_indices], "x", color='red', markersize=8, label='Peaks')
        plt.title(f"Peak Detection in drive05")
        plt.xlabel("Sample Index")
        plt.ylabel(column_name)
        plt.legend()
        plt.grid(True)
        plt.show()

    return peak_indices

detected_peaks = detect_marker_peaks(
    csv_path="./DRIVEDB/drive_csv/drive05.csv",
    distance_min=800,
    prominence_min=2,
    show_plot=True
)

if len(detected_peaks) > 0:
    print(f"{detected_peaks}")
```

```python
def normalize_columns(df):
    cols = {c: c for c in df.columns}
    for c in df.columns:
        cl = c.strip().lower()
        if cl.startswith("ecg"):
            cols[c] = "ECG-mV"
        elif ("hand gsr" in cl) or ("hand" in cl):
            cols[c] = "hand GSR-mV"
        elif cl.startswith("resp"):
            cols[c] = "RESP-mV"
        elif "marker" in cl:
            cols[c] = "marker"
        elif cl.startswith("hr"):
            cols[c] = "HR-bpm"
    return df.rename(columns=cols)

def get_ends_for_drive(marker_df, drive_name):
    r = marker_df[marker_df["Driver"] == drive_name.lower()]
    if r.empty:
        return None
    row = r.iloc[0]
    ends = []
    for c in SEG_ORDER:
        v = row.get(c, np.nan)
        if pd.isna(v):
            ends.append(None)
        else:
            ends.append(int(v))
    return ends

def build_segments_from_ends(ends, n_samples):
    segs = []
    prev = 0
    for i, col in enumerate(SEG_ORDER):
        e = ends[i]
        end_i = int(e) if (e is not None and not pd.isna(e)) else n_samples
        start_i = int(prev)
        start_i = max(0, min(start_i, n_samples))
        end_i = max(0, min(end_i, n_samples))
        if end_i > start_i:
            segs.append({
                "segment_raw": col,
                "segment": RAW_TO_SHORT.get(col, "unknown"),
                "start_i": start_i,
                "end_i": end_i
            })
        prev = end_i if end_i > prev else prev
    return segs

def label_samples(df_signal, segments):
    n = len(df_signal)
    labels_bin = np.array(["unknown"] * n, dtype=object)
    labels_3 = np.array(["unknown"] * n, dtype=object)
    for seg in segments:
        s = max(0, int(seg["start_i"]))
        e = min(n, int(seg["end_i"]))
        if e <= s:
            continue
        lab = seg["segment"]
        labels_bin[s:e] = BINARY_MAP.get(lab, "unknown")
        labels_3[s:e]  = THREE_MAP.get(lab, "unknown")
    output = df_signal.copy()
    output["time_s"] = np.arange(n) / FS
    output["label_binary"] = labels_bin
    output["label_3class"] = labels_3
    return output
```

```python
csv_files = sorted(list(RAW_CSV_DIR.glob("drive*.csv")))

for f in csv_files:
    drive = f.stem.lower()
    print("Processing", drive)

    df = pd.read_csv(f)
    df = normalize_columns(df)
    n = len(df)
    duration_s = n / FS

    ends = get_ends_for_drive(marker_df, drive)
    if ends is None:
        print("No marker_info entry for", drive, "- skipping")
        continue

    segments = build_segments_from_ends(ends, n)

    seg_lens_s = [(seg["end_i"]-seg["start_i"])/FS for seg in segments]
    min_seg = min(seg_lens_s) if seg_lens_s else 0
    if min_seg < MIN_SEG_SEC:
        print(f"WARNING: very short segment (<{MIN_SEG_SEC}s)")

    df_sample = label_samples(df, segments)

    last_known_end_i = 0
    if segments:
        last_known_end_i = segments[-1]["end_i"]

    if last_known_end_i > 0 and last_known_end_i < len(df_sample):
        samples_to_drop = len(df_sample) - last_known_end_i
        df_sample = df_sample.iloc[:last_known_end_i]

    df_sample.to_csv(LABELED_DIR / f"{drive}_labeled.csv", index=False)
```

## II.2 Filtering

```python
import numpy as np
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft, fftfreq
from scipy.signal import butter, filtfilt

ECG_CUTOFF_HZ = 0.5
RESP_LOWPASS_CUTOFF = 0.3
RESP_FIR_ORDER = 10
MOVING_AVG_WIDTH = 5

FILTERED_DIR = Path("./filtered_output")
FILTERED_DIR.mkdir(parents=True, exist_ok=True)
```

```python
def fft_remove_baseline(signal, fs, cutoff_hz):
    N = len(signal)
    freqs = fftfreq(N, d=1/fs)
    spectrum = fft(signal)
    spectrum[np.abs(freqs) < cutoff_hz] = 0
    filtered = np.real(ifft(spectrum))
    return filtered

def fft_remove_dc(signal, fs):
    N = len(signal)
    spectrum = fft(signal)
    spectrum[0] = 0
    return np.real(ifft(spectrum))

def filter_resp_signal(signal, fs):
    nyq = fs / 2
    Wn = 0.05 / nyq
    b, a = butter(4, Wn, btype='highpass')

    return filtfilt(b, a, signal)
```

```python
def apply_filter_sliding(signal, fs, window_sec, step_sec, mode="ecg", cutoff_hz=0.5):
    window_len = int(window_sec * fs)
    step_len = int(step_sec * fs)
    n = len(signal)

    filtered = np.zeros(n)
    count = np.zeros(n)

    for start in range(0, n - window_len + 1, step_len):
        end = start + window_len
        seg = signal[start:end]

        if mode == "ecg":
            cleaned = fft_remove_baseline(seg, fs, cutoff_hz)
        elif mode == "gsr":
            cleaned = fft_remove_dc(seg, fs)
        elif mode == "resp":
            cleaned = filter_resp_signal(seg, fs)

        filtered[start:end] += cleaned
        count[start:end] += 1

    count[count == 0] = 1
    return filtered / count
```

```python
def process_all_csv_files():
    csv_files = sorted(LABELED_DIR.glob("drive*_labeled.csv"))

    for f in csv_files:
        df = pd.read_csv(f)

        if ECG_COL in df.columns:
            df[ECG_COL] = apply_filter_sliding(df[ECG_COL].values, FS, WINDOW_SEC, STEP_SEC, mode="ecg", cutoff_hz=ECG_CUTOFF_HZ)

        if FGSR_COL in df.columns:
            df[FGSR_COL] = apply_filter_sliding(df[FGSR_COL].values, FS, WINDOW_SEC, STEP_SEC, mode="gsr")

        if HGSR_COL in df.columns:
            df[HGSR_COL] = apply_filter_sliding(df[HGSR_COL].values, FS, WINDOW_SEC, STEP_SEC, mode="gsr")

        if RESP_COL in df.columns:
            df[RESP_COL] = apply_filter_sliding(df[RESP_COL].values, FS, WINDOW_SEC, STEP_SEC, mode="resp")

        new_filename = f.name.replace("_labeled.csv", "_filtered.csv")
        out_path = FILTERED_DIR / new_filename
        df.to_csv(out_path, index=False)
        print(f"{out_path.name}")

process_all_csv_files()
```

```python
from scipy.signal import find_peaks

def plot_ecg_peaks(filtered_dir, filename, ecg_col="ECG-mV",
                   fs=15.5, t0=0, window_sec=60):
    fpath = Path(filtered_dir) / filename
    df = pd.read_csv(fpath)

    ecg = df[ecg_col].values
    time = df["time_s"].values

    t1 = t0 + window_sec
    mask = (time >= t0) & (time < t1)
    ecg_seg = ecg[mask]
    time_seg = time[mask]

    distance = 5
    peaks, props = find_peaks(
        ecg_seg,
        distance=distance,
        prominence=0.1
    )

    print(f"{filename} | [{t0}s–{t1}s]: {len(peaks)} peaks")

    plt.figure(figsize=(15, 4))
    plt.plot(time_seg, ecg_seg, label="ECG raw")
    plt.xlabel("Time (s)")
    plt.ylabel("ECG (mV)")
    plt.title(f"ECG: {filename}")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

plot_ecg_peaks(LABELED_DIR, "drive05_labeled.csv", t0=0, window_sec=60)
```

# IV. Feature Extraction

```python
import pandas as pd
import numpy as np
from pathlib import Path
from scipy.signal import find_peaks, savgol_filter

MIXED_THRESH = 0.2
HR_MAX_VALID = 180.0

LABEL_BINARY_COL = "label_binary"
LABEL_3CLASS_COL = "label_3class"
TIME_COL = "time_s"

SEGMENT_OUTPUT_DIR = Path("./segment_output")
SEGMENT_OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

FEATURES_DIR = Path("./features_output")
FEATURES_DIR.mkdir(parents=True, exist_ok=True)

BAD_WINDOWS_DICT = {}

def get_label_at_center(labels, time_values, start_s, end_s):
    center_time = (start_s + end_s) / 2.0
    center_idx = np.abs(time_values - center_time).argmin()
    return labels[center_idx]

def segment_drive(file_path, window_sec=WINDOW_SEC, step_sec=STEP_SEC):
    df = pd.read_csv(file_path)
    if TIME_COL not in df.columns or LABEL_BINARY_COL not in df.columns or LABEL_3CLASS_COL not in df.columns:
        print(f"Missing required columns in {file_path.name}")
        return pd.DataFrame()

    time_values = df[TIME_COL].values
    total_duration = time_values[-1]
    starts = np.arange(0, total_duration - window_sec + 1e-6, step_sec)

    segments = []
    bad_indices = []

    labels_bin = df[LABEL_BINARY_COL].values
    labels_3c = df[LABEL_3CLASS_COL].values

    for i, s in enumerate(starts):
        e = s + window_sec
        mask = (time_values >= s) & (time_values < e)
        idx = np.where(mask)[0]

        if len(idx) < 1:
            continue

        label_bin_center = get_label_at_center(labels_bin, time_values, s, e)
        label_3c_center = get_label_at_center(labels_3c, time_values, s, e)

        prop_mismatch = np.mean(labels_bin[idx] != label_bin_center)
        if prop_mismatch > MIXED_THRESH:
            bad_indices.append(i)

        segments.append({
            "window_index": i,
            "window_start_s": s,
            "window_end_s": e,
            "n_samples": len(idx),
            "label_binary": label_bin_center,
            "label_3class": label_3c_center,
            "mismatch_ratio": prop_mismatch
        })

    BAD_WINDOWS_DICT[file_path.name] = bad_indices

    return pd.DataFrame(segments)

def run_segment_pipeline():
    all_csv = sorted(FILTERED_DIR.glob("drive*_filtered.csv"))
    for f in all_csv:
        df_seg = segment_drive(f)
        if df_seg is not None and not df_seg.empty:
            out_path = SEGMENT_OUTPUT_DIR / f"{f.stem.replace('_filtered','')}_segmented.csv"
            df_seg.to_csv(out_path, index=False)
            print(f"{out_path.name} ({len(df_seg)} windows)")
        else:
            print(f"No valid windows for {f.name}")

run_segment_pipeline()
```

### ECG

```python
def extract_ecg_features(ecg_signal, fs):
    peaks, _ = find_peaks(ecg_signal, distance=5, prominence=0.1)

    num_peaks = len(peaks)
    if num_peaks > 1:
        rr_intervals = np.diff(peaks) / fs
        mean_rr = np.mean(rr_intervals)
    else:
        mean_rr = np.nan

    return {
        "ECG_peaks":num_peaks,
        "mean_rr_interval": mean_rr
    }
```

### EDA (fGSR & hGSR)

```python
def extract_gsr_features(gsr_signal, fs):
    if len(gsr_signal) < 5:
        return np.nan, np.nan, np.nan, 0

    peaks, _ = find_peaks(gsr_signal, distance=5, prominence=0.1)
    num_peaks = len(peaks)

    if num_peaks >= 2:
        peak_intervals = np.diff(peaks) / fs
        peak_values = gsr_signal[peaks]
        peak_diffs = np.abs(np.diff(peak_values))

        mean_peak_interval = np.mean(peak_intervals)
        mean_peak_diff = np.mean(peak_diffs)
        has_peaks = 1
    else:
        mean_peak_interval = float(WINDOW_SEC)
        mean_peak_diff = 0.0
        has_peaks = 0

    return num_peaks, mean_peak_interval, mean_peak_diff, has_peaks
```

### RESP

```python
def extract_resp_features(resp_signal, fs):
    if len(resp_signal) < 5:
        return np.nan

    peaks, _ = find_peaks(resp_signal, distance=5, prominence=0.1)

    if len(peaks) >= 2:
        rr_intervals = np.diff(peaks) / fs
        mean_rr_resp = np.mean(rr_intervals)
    else:
        mean_rr_resp = np.nan

    return mean_rr_resp
```

EMG

```python
import numpy as np

def extract_emg_features(emg_signal, fs):
    if len(emg_signal) < 5:
        return np.nan

    rms_emg = np.sqrt(np.mean(emg_signal**2))

    return np.round(rms_emg, 6)
```

### HR

```python
def extract_hr_features(hr_signal):
    if len(hr_signal) < 2 or np.all(np.isnan(hr_signal)):
        return {
            "HR_mean": np.nan,
            "HR_std": np.nan,
            "HR_rmssd": np.nan,
            "HR_min": np.nan,
            "HR_max": np.nan
        }

    diffs = np.diff(hr_signal)
    rmssd = np.sqrt(np.mean(diffs**2))

    return {
        "HR_mean": np.mean(hr_signal),
        "HR_std": np.std(hr_signal),
        "HR_rmssd": rmssd,
        "HR_min": np.min(hr_signal),
        "HR_max": np.max(hr_signal)
    }
```

### Batch

```python
def get_label_at_center(labels, time_values, start_s, end_s):
    center_time = (start_s + end_s) / 2.0
    center_idx = np.abs(time_values - center_time).argmin()
    return labels[center_idx]

def extract_features_for_drive(csv_path, fs, window_sec, step_sec):
    df = pd.read_csv(csv_path)

    required_cols = ["time_s", "ECG-mV", "foot GSR-mV", "hand GSR-mV", "RESP-mV", "HR-bpm", "label_binary", "label_3class"]
    for col in required_cols:
        if col not in df.columns:
            raise ValueError(f"Thiếu cột {col} trong file {csv_path}.")

    drive_id = csv_path.stem.split("_filtered")[0]

    time_arr = df["time_s"].values
    ecg_arr = df["ECG-mV"].values
    fgsr_arr = df["foot GSR-mV"].values
    hgsr_arr = df["hand GSR-mV"].values
    resp_arr = df["RESP-mV"].values
    emg_arr = df["EMG-mV"].values
    hr_arr = df["HR-bpm"].values
    lbl_bin_arr = df["label_binary"].values
    lbl_3c_arr = df["label_3class"].values

    total_duration = time_arr[-1]
    window_starts = np.arange(0, total_duration - window_sec + 1, step_sec)
    results = []

    for start in window_starts:
        end = start + window_sec

        mask = (time_arr >= start) & (time_arr < end)
        indices = np.where(mask)[0]

        expected_samples = int(window_sec * fs)
        if len(indices) < (expected_samples * 0.8):
            continue

        label_binary_center = get_label_at_center(lbl_bin_arr, time_arr, start, end)
        label_3class_center = get_label_at_center(lbl_3c_arr, time_arr, start, end)

        window_labels_bin = lbl_bin_arr[indices]
        prop_mismatch = np.mean(window_labels_bin != label_binary_center)
        if prop_mismatch > MIXED_THRESH:
            continue

        ecg_seg = ecg_arr[indices]
        fgsr_seg = fgsr_arr[indices]
        hgsr_seg = hgsr_arr[indices]
        resp_seg = resp_arr[indices]
        emg_seg = emg_arr[indices]
        hr_seg = hr_arr[indices]

        valid_hr_mask = (hr_seg > 0) & (hr_seg <= HR_MAX_VALID)
        hr_seg_clean = hr_seg[valid_hr_mask]

        if len(hr_seg_clean) < (len(hr_seg) * 0.5):
            continue

        if np.min(hr_seg_clean) == 0:
             continue

        hr_feat = extract_hr_features(hr_seg_clean)

        if 'HR_min' in hr_feat and (hr_feat['HR_min'] == 0 or np.isnan(hr_feat['HR_min'])):
            continue

        hr_max = hr_feat.get('HR_max', 0)

        if (label_binary_center in "no" or label_3class_center in "low") and (hr_max > 100):
            continue

        ecg_feat = extract_ecg_features(ecg_seg, fs)

        f_peaks, f_int, f_diff, f_flag = extract_gsr_features(fgsr_seg, fs)
        h_peaks, h_int, h_diff, h_flag = extract_gsr_features(hgsr_seg, fs)

        resp_mean_rr = extract_resp_features(resp_seg, fs)

        emg_rms = extract_emg_features(emg_seg, fs)

        features = {
            "drive_id": drive_id,
            "start_s": start,
            "end_s": end,

            **ecg_feat,

            "fGSR_peaks": f_peaks,
            "fGSR_mean_peak_interval": f_int,
            "fGSR_mean_peak_diff": f_diff,
            "fGSR_has_peaks": f_flag,

            "hGSR_peaks": h_peaks,
            "hGSR_mean_peak_interval": h_int,
            "hGSR_mean_peak_diff": h_diff,
            "hGSR_has_peaks": h_flag,

            "RESP_mean_rr": resp_mean_rr,

            "EMG_rms": emg_rms,

            **hr_feat,

            "label_binary": label_binary_center,
            "label_3class": label_3class_center,

            # "mismatch_ratio": prop_mismatch
        }

        results.append(features)

    return pd.DataFrame(results)
```

```python
def process_all_drives_in_folder():
    all_csv = sorted(FILTERED_DIR.glob("drive*_filtered.csv"))

    for f in all_csv:
        print(f"Processing: {f.name}")
        df_feat = extract_features_for_drive(f, fs=FS, window_sec=WINDOW_SEC, step_sec=STEP_SEC)

        out_path = FEATURES_DIR / f"{f.stem.replace('_filtered','')}_features.csv"
        df_feat.to_csv(out_path, index=False)
        print(f"Saved: {out_path.name}")

process_all_drives_in_folder()
```

```python
import pandas as pd
import numpy as np
from pathlib import Path
from sklearn.model_selection import GroupKFold, LeaveOneGroupOut
from sklearn.preprocessing import LabelEncoder, StandardScaler

def load_and_prepare_cv(features_dir, target_col="label_binary"):
    data_path = Path(features_dir)
    csv_files = sorted(list(data_path.glob("*.csv")))

    full_df = []
    for f in csv_files:
        temp_df = pd.read_csv(f)
        full_df.append(temp_df)

    full_df = pd.concat(full_df, ignore_index=True)
    full_df = full_df[full_df[target_col] != 'unknown'].reset_index(drop=True)

    ignore_cols = [
        'drive_id', 'start_s', 'end_s',
        'label_binary', 'label_3class',
        'fGSR_has_peaks', 'hGSR_has_peaks'
    ]

    numeric_feats = [c for c in full_df.columns
                     if c not in ignore_cols
                     and np.issubdtype(full_df[c].dtype, np.number)]
    print(numeric_feats)

    flag_feats = [c for c in ['fGSR_has_peaks', 'hGSR_has_peaks'] if c in full_df.columns]

    final_features = numeric_feats + flag_feats

    X = full_df[final_features].values
    y = full_df[target_col].values
    groups = full_df['drive_id'].values

    if target_col == 'label_3class':
        y_mapped = full_df[target_col].map({'low': 0, 'medium': 1, 'high': 2})
        if y_mapped.isnull().any():
             y = LabelEncoder().fit_transform(y)
        else:
             y = y_mapped.values
             LabelEncoder().classes_ = np.array(['low', 'medium', 'high'])
    else:
        y_raw_labels = full_df["label_binary"].astype(str).str.lower().values

        y_mapped = pd.Series(y_raw_labels).map({'no': 0, 'yes': 1})

        full_df = full_df[~y_mapped.isnull()].reset_index(drop=True)
        X = full_df[final_features].values
        groups = full_df['drive_id'].values
        y = y_mapped.dropna().values.astype(int)

    unique_labels, counts = np.unique(y, return_counts=True)

    label_names = ['No Stress (0)', 'Stress (1)'] if len(unique_labels) == 2 else [f'Class {l}' for l in unique_labels]

    print("-" * 50)
    print(f"📊 THỐNG KÊ NHÃN ({target_col.upper()})")
    print("-" * 50)

    total = len(y)

    for label, count in zip(unique_labels, counts):
        name = label_names[label] if label < len(label_names) else f'Unknown Label ({label})'
        print(f"✅ {name}: {count} mẫu ({count/total:.2%})")

    print(f"Tổng số mẫu: {total}")
    print("-" * 50)

    n_drivers = len(np.unique(groups))
    unique_drivers = np.unique(groups)

    if n_drivers < 10:
        cv_splitter = LeaveOneGroupOut()
        print(f"eave-One-Group-Out ({n_drivers} folds)")
    else:
        cv_splitter = GroupKFold(n_splits=10)
        print(f"10-Fold GroupKFold")

    return X, y, groups, cv_splitter, numeric_feats, flag_feats
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import LeaveOneGroupOut, GroupKFold

def train_and_evaluate_full_metrics_en(features_dir, task_type="binary", model_type="rf"):
    target_col = "label_binary" if task_type == "binary" else "label_3class"
    print(f"\n{'='*60}")
    print(f"📊 EVALUATING MODEL LAYER: {task_type.upper()} ({target_col})")
    print(f"{'='*60}")

    data_pack = load_and_prepare_cv(features_dir, target_col=target_col)

    if data_pack is None or data_pack[0] is None:
        print("Error: Could not load data.")
        return None

    X, y, groups, cv, num_cols, flag_cols = data_pack

    unique_labels = np.unique(y)

    metrics_list = []
    accumulated_cm = None

    for i, (train_idx, test_idx) in enumerate(cv.split(X, y, groups)):
        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]

        scaler = StandardScaler()
        n_num = len(num_cols)
        X_train_num = scaler.fit_transform(X_train[:, :n_num])
        X_test_num = scaler.transform(X_test[:, :n_num])

        if len(flag_cols) > 0:
            X_train_final = np.hstack((X_train_num, X_train[:, n_num:]))
            X_test_final = np.hstack((X_test_num, X_test[:, n_num:]))
        else:
            X_train_final = X_train_num
            X_test_final = X_test_num

        print(flag_cols)

        if model_type == "rf":
            model = RandomForestClassifier(n_estimators=110,
                                           random_state=1,
                                           criterion='entropy',
                                           class_weight='balanced')

        model.fit(X_train_final, y_train)
        y_pred = model.predict(X_test_final)

        avg_method = 'binary' if task_type == 'binary' else 'weighted'
        pos_label = 1 if len(unique_labels) == 2 else None

        acc = accuracy_score(y_test, y_pred)
        prec = precision_score(y_test, y_pred, average=avg_method, pos_label=pos_label, zero_division=0)
        rec = recall_score(y_test, y_pred, average=avg_method, pos_label=pos_label, zero_division=0)
        f1 = f1_score(y_test, y_pred, average=avg_method, pos_label=pos_label, zero_division=0)

        cm = confusion_matrix(y_test, y_pred, labels=[1, 0])

        spec = np.nan
        if len(unique_labels) == 2:
            tn, fp, fn, tp = cm.ravel()
            spec = tn / (tn + fp) if (tn + fp) > 0 else 0.0

        if accumulated_cm is None:
            accumulated_cm = cm
        else:
            if cm.shape == accumulated_cm.shape:
                accumulated_cm += cm

        test_drivers = np.unique(groups[test_idx])
        metrics_list.append({
            'Fold': i + 1,
            'Test_Drivers': str(test_drivers),
            'Accuracy': acc,
            'Precision': prec,
            'Recall (Sensitivity)': rec,
            'Specificity': spec,
            'F1_Score': f1
        })

        print(f"Fold {i+1:02d} | Test: {test_drivers} | Acc: {acc:.4f} | Sens: {rec:.4f} | Spec: {spec:.4f} | F1: {f1:.4f}")

    results_df = pd.DataFrame(metrics_list)
    print("\n" + "-"*80)
    print("📈 AVERAGE PERFORMANCE METRICS:")
    print("-" * 80)
    print(results_df[['Accuracy', 'Precision', 'Recall (Sensitivity)', 'Specificity', 'F1_Score']].mean().to_string())
    print(f"\nStd Dev of Accuracy: {results_df['Accuracy'].std():.4f}")
    print("-" * 80)

    plt.figure(figsize=(6, 5))

    if task_type == 'binary':
        labels = ['Stress (1)', 'No Stress (0)']
    elif task_type == '3class':
        labels = ['Low', 'Medium', 'High']
    else:
        labels = ['Medium', 'High']

    ax = sns.heatmap(accumulated_cm, annot=True, fmt='d', cmap='Blues',
                     xticklabels=labels, yticklabels=labels)
    ax.xaxis.tick_top()
    ax.xaxis.set_label_position('top')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.title(f'Accumulated Confusion Matrix', y=1.15)

    plt.tight_layout()
    plt.show()

    return results_df

results_df = train_and_evaluate_full_metrics_en(FEATURES_DIR, task_type="binary", model_type="rf")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.model_selection import LeaveOneGroupOut, GroupKFold

WINDOW_SEC = 60
STEP_SEC = 30
SEQ_LEN = int(WINDOW_SEC * FS)
MIXED_THRESH = 0.2
HR_INTERPOLATE_LIMIT_SEC = 5
CHANNELS = ["ECG-mV", "foot GSR-mV", "hand GSR-mV", "RESP-mV", "HR-bpm", "EMG-mV"]
LOW_STRESS_LABELS = ['low', 'no']


CNN1_FILTERS = 32
CNN1_KERNEL = 10
CNN2_FILTERS = 32
CNN2_KERNEL = 5
CNN_STRIDES = 2
POOL_SIZE = 2


DROPOUT_RATE = 0.3
LSTM1_UNITS = 48
LSTM2_UNITS = 24
DENSE_UNITS = 32

LEARNING_RATE = 0.0001
BATCH_SIZE = 32
PATIENCE = 25
EPOCHS = 50
```

```python
def get_label_at_center(labels, time_values, start_s, end_s):
    center_time = (start_s + end_s) / 2.0
    center_idx = np.abs(time_values - center_time).argmin()
    return labels[center_idx]

def data_segment(input_folder):
    input_path = Path(input_folder)
    csv_files = sorted(list(input_path.glob("*.csv")))

    X_seq, y_seq, groups = [], [], []

    for f in csv_files:
        try:
            df = pd.read_csv(f)

            # Xử lý HR (Nội suy có giới hạn)
            if 'HR-bpm' in df.columns:
                df['HR-bpm'] = df['HR-bpm'].replace(0, np.nan)
                limit_samples = int(HR_INTERPOLATE_LIMIT_SEC * FS)
                df['HR-bpm'] = df['HR-bpm'].interpolate(method='linear', limit=limit_samples)

            # --- BASELINE NORMALIZATION ---
            df_baseline = df[df['label_3class'].isin(LOW_STRESS_LABELS)].copy()

            if len(df_baseline) < 50: continue

            baseline_mean = df_baseline[CHANNELS].mean()
            baseline_std = df_baseline[CHANNELS].std()
            baseline_std[baseline_std == 0] = 1e-6

            # Chuẩn hóa toàn bộ dữ liệu DF theo Baseline Stats
            df[CHANNELS] = (df[CHANNELS] - baseline_mean) / baseline_std

            time_arr = df["time_s"].values
            signal_arr_norm = df[CHANNELS].values
            label_arr = df["label_3class"].values
            drive_id = f.stem.split('_')[0]

            total_duration = time_arr[-1]
            starts = np.arange(0, total_duration - WINDOW_SEC + 1, STEP_SEC)

            for s in starts:
                e = s + WINDOW_SEC

                idx_start = np.searchsorted(time_arr, s)
                idx_end = np.searchsorted(time_arr, e)

                segment_scaled = signal_arr_norm[idx_start:idx_end]

                if len(segment_scaled) != SEQ_LEN: continue
                if np.isnan(segment_scaled).any(): continue

                center_label = get_label_at_center(label_arr, time_arr, s, e)

                if center_label not in ['medium', 'high']: continue

                window_labels = label_arr[idx_start:idx_end]
                if np.mean(window_labels != center_label) > MIXED_THRESH: continue

                X_seq.append(segment_scaled)
                y_seq.append(center_label)
                groups.append(drive_id)

        except Exception as e:
            continue

    if len(X_seq) == 0:
        print("Không tạo được mẫu dữ liệu nào.")
        return None, None, None

    # Mapping thủ công: Medium -> 0, High -> 1
    mapping_rule = {'medium': 0, 'high': 1}
    y_enc = np.array([mapping_rule[lbl] for lbl in y_seq])

    print(f"Dataset Shape: {np.array(X_seq).shape}")

    return np.array(X_seq), y_enc, np.array(groups)
```

```python
import tensorflow as tf

def build_model(input_shape):
    model = tf.keras.Sequential(name="CNN_LSTM")

    model.add(tf.keras.layers.Input(shape=input_shape))

    model.add(tf.keras.layers.Conv1D(
        filters=CNN1_FILTERS,
        kernel_size=CNN1_KERNEL,
        strides=CNN_STRIDES,
        padding='valid'))
    model.add(tf.keras.layers.ELU(alpha=0.1))
    model.add(tf.keras.layers.BatchNormalization())
    model.add(tf.keras.layers.Dropout(DROPOUT_RATE))
    model.add(tf.keras.layers.MaxPooling1D(
        pool_size=POOL_SIZE,
        strides=POOL_SIZE))

    model.add(tf.keras.layers.Conv1D(
        filters=CNN2_FILTERS,
        kernel_size=CNN2_KERNEL,
        strides=CNN_STRIDES,
        padding='valid'))
    model.add(tf.keras.layers.ELU(alpha=0.1))
    model.add(tf.keras.layers.BatchNormalization())
    model.add(tf.keras.layers.Dropout(DROPOUT_RATE))
    model.add(tf.keras.layers.MaxPooling1D(
        pool_size=POOL_SIZE,
        strides=POOL_SIZE))

    model.add(tf.keras.layers.LSTM(LSTM1_UNITS, return_sequences=True))
    model.add(tf.keras.layers.Dropout(DROPOUT_RATE))

    model.add(tf.keras.layers.LSTM(LSTM2_UNITS, return_sequences=False))
    model.add(tf.keras.layers.Dropout(DROPOUT_RATE))

    model.add(tf.keras.layers.Dense(DENSE_UNITS, activation='relu'))
    model.add(tf.keras.layers.Dense(1, activation='sigmoid'))

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=LEARNING_RATE),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model
```

```python
def run_model(raw_input_folder):
    X, y, groups = data_segment(raw_input_folder)

    if X is None: return

    unique_groups = np.unique(groups)
    sorted_groups = sorted(unique_groups)

    metrics_list = []
    accumulated_cm = None
    fold = 1

    print("\n" + "="*80)
    print("TRAINING OPTIMIZED 1D CNN-LSTM (SEQUENTIAL CV)")
    print("="*80)

    for test_driver_id in sorted_groups:

        test_idx = np.where(groups == test_driver_id)[0]
        train_idx = np.where(groups != test_driver_id)[0]

        if len(test_idx) == 0 or len(train_idx) == 0:
            continue

        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]

        model = build_model((X.shape[1], X.shape[2]))

        early_stop = tf.keras.callbacks.EarlyStopping(
            monitor='val_loss', patience=PATIENCE, restore_best_weights=True, verbose=0
        )

        print(f"Fold {fold:02d} (Test: ['{test_driver_id}'])...", end="")

        model.fit(X_train, y_train, validation_data=(X_test, y_test),
                  epochs=EPOCHS, batch_size=BATCH_SIZE, callbacks=[early_stop], verbose=0)

        y_pred = (model.predict(X_test, verbose=0) > 0.5).astype(int).flatten()

        acc = accuracy_score(y_test, y_pred)
        prec = precision_score(y_test, y_pred, pos_label=1, zero_division=0)
        sens = recall_score(y_test, y_pred, pos_label=1, zero_division=0)
        f1 = f1_score(y_test, y_pred, pos_label=1, zero_division=0)

        cm = confusion_matrix(y_test, y_pred, labels=[1, 0])
        tn, fp, fn, tp = cm.ravel()
        spec = tn / (tn + fp) if (tn + fp) > 0 else 0.0

        if accumulated_cm is None: accumulated_cm = cm
        else: accumulated_cm += cm

        print(f" -> Acc={acc:.4f} | Sens={sens:.4f} | Spec={spec:.4f} | F1={f1:.4f}")

        metrics_list.append({'Accuracy': acc, 'Sensitivity': sens, 'Specificity': spec, 'Precision': prec, 'F1_Score': f1})
        fold += 1

    results_df = pd.DataFrame(metrics_list)
    print("\n" + "="*80)
    print("AVERAGE PERFORMANCE METRICS:")
    print("-" * 80)
    print(results_df[['Accuracy', 'Sensitivity', 'Specificity', 'Precision', 'F1_Score']].mean().to_string())
    print(f"\nStd Dev of Accuracy: {results_df['Accuracy'].std():.4f}")

    plt.figure(figsize=(6, 5))
    ax = sns.heatmap(accumulated_cm, annot=True, fmt='d', cmap='Oranges',
                 xticklabels=['High (1)', 'Medium (0)'],
                 yticklabels=['High (1)', 'Medium (0)'])
    ax.xaxis.tick_top()

    ax.xaxis.set_label_position('top')

    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

    return results_df

run_model(FILTERED_DIR)
```
